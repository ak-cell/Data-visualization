```python
import matplotlib.pyplot as plt
```

```python
#line graph
x=[1,2,3]
y=[4,5,2]
plt.plot(x,y)
plt.show()
```

```python
#Adding title and axis name
x=[1,2,3]
y=[3,6,9]
plt.plot(x,y)
plt.title("New graph")
plt.xlabel("distance")
plt.ylabel("time")
plt.show()
```

```python


from matplotlib import style
style.use('ggplot')
x1=[1,2,3]
y1=[4,5,2]

x2=[6,7,9]
y2=[8,10,12]
plt.plot(x1,y1,'g',label='first',linewidth=5)
plt.plot(x2,y2,'k',label='second',linewidth=5)
plt.title('New Graph')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.grid(True,color="b")
plt.show()
```

```python
# Bar graph
x1=[1,3,5,7]
y1=[5,2,6,6]
x2=[2,4,6,8]
y2=[6,3,5,2]
plt.bar(x1,y1,label="first bar chart")
plt.bar(x2,y2,label="second bar chart")
plt.legend()
plt.xlabel("bar number")
plt.ylabel("bar height")
plt.title("bar chart")
plt.show()

```

```python
# Histogram
sale_value=[12,10,5,7,0,14,15,24,10,11,15,30,18,11,10,18,30]
bins=[0,5,10,15,20,25,30]
plt.hist(sale_value,bins,histtype='bar',rwidth=0.6)
plt.xlabel('x axis')
plt.ylabel("y axis")
plt.title("Histogram")
plt.show()
```

```python
#Scatterplot
x=[1,2,3,4,8,5,6]
y=[3,5,6,7,8,9,3]
plt.scatter(x,y,label="scatters",color="violet")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scatter plot")
plt.show()
```

```python
#Stackplot
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]
plt.stackplot(days,sleeping,eating,working,playing,colors=("red","green","blue","yellow"))
plt.xlabel("x")
plt.ylabel("y")
plt.title("interesting graph\n Check it out")
plt.legend()
plt.show()
```

```python
#Pie Chart
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['yellow','lightgreen','pink','skyblue']
explode=[0.1,0,0,0]  #explode 1st slice
#Plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True,startangle=140)
plt.axis('equal')
plt.show()
```
